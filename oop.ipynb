```python
class Coche:
    tipo = "vehiculo"

    def __init__(self,modelo,posicion=0):
        self.modelo=modelo
        self.posicion= posicion
    
    def acelerar(self,metros):
        self.posicion = metros + self.posicion
        print(f"{self.modelo} acelera y avanza {metros}m. Nueva posicion: {self.posicion} ")
    
    def girar(self,direccion):
        print(f"{self.modelo} gira a la {direccion}")

    def frenar(self):
        print(f"{self.modelo} frena y se detiene en la posicion {self.posicion}")

mi_coche = Coche("ranger")
mi_coche.acelerar(50)
mi_coche.frenar()
mi_coche.girar("derecha")
```

Challenge 🤺15 min
Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

Observación: Editar las clases hijas para que no se inicialicen.

```python
class Gato:
    tipo="Gato"

    def __init__(self,sonido,comer):
        self.sonido=sonido
        self.comer=comer
    def hablar(self):
        print(f"el {self.tipo} dice {self.sonido}")
    def hambre(self):
        print(f"el {self.tipo} {self.comer}")

class Anaranjado(Gato):
    def comportamiento(self):
        print(f"el {self.tipo} anarnjado tiene mucha mucha energia")

class Blanco(Gato):
    def dormir(self):
        print(f"el {self.tipo} blanco duerme mucho")

loco = Anaranjado("miau","come mucho")
loco.hablar()
loco.hambre()
loco.comportamiento()

kaigue=Blanco("miau","como mucho")
kaigue.hablar()
kaigue.hambre()
kaigue.dormir()
```

Challenge 🤺
Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un hereo y un villano.
Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

Ejemplo: Yo me llamo pepito y tengo super fuerza
Ejemplo: Yo me llamo juanito y puedo volar

```python
class Personajes:

    def __init__(self,nombre,bueno_malo):
        self.nombre=nombre
        self.bueno_malo=bueno_malo
    def presentarse(self):
        print(f"yo soy el {self.nombre} y soy el {self.bueno_malo}")
class Heroe(Personajes):
    def volar(self,poder_bueno):
        self.poder_bueno=poder_bueno
        print(f"yo puedo {self.poder_bueno}")
class Villano(Personajes):
    def romper_paredes(self,poder_malo):
        self.poder_malo=poder_malo
        print(f"yo puedo {poder_malo}")
        

bueno= Heroe("juan","heroe")
bueno.presentarse()
bueno.volar("volar")

malo=Villano("jose","villano")
malo.presentarse()
malo.romper_paredes("romper paredes")
```

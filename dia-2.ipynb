Challenge 🤺 - 30m
Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

Entre -10 y 13 grados, que imprima "Ro'y koko".
Entre 14 y 23 grados, que imprima "Refresca'i".
Entre 24 y 28 grados, que imprima "Oikoite".
Entre 29 y 45 grados, que imprima "Hakurooooo".
Para todo lo demás, imprimir "Ñamano mba'e!".

```python
temp = int(input("introduce la temperatura (en grados Celsius)"))

if temp >= -10 and temp <= 13:
    print("Ro'y koko")
elif temp >= 14 and temp <= 23:
    print("Refresca'i")
elif temp >= 24 and temp <= 28:
    print("Oikoite")
elif temp >= 29 and temp <= 45:
    print("Hakurooooo")
else:
    print("Ñamano mba'e!")
```

Challenge 🤺 - 20m
Escribir un programa que almacene una contraseña en una variable, y pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.

Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

```python
contra=1234

while True:
    contrasenha=int(input("Introduce la contrasenha"))
    if contrasenha == contra:
        print("contrasenha correcta")
        break
    else:
        print("contrasenha incorrecta, intente nuevamente")
```

Crear una lista de 5 elementos. Luego, seguir los siguientes pasos.

Challenge 🤺 - 20m
Recorrer los elementos de una lista de marcas e imprimir cada marca solamente si tiene 6 letras o más.

```python
marcas = ["nike", "adidas","reebok","under armour","puma","new balance"]

for marca in marcas:
    if len(marca)>= 6:
        print(marca)
```

Challenge 🤺 - 15m
Solicitar al usuario que ingrese una frase.

Imprimir un mensaje cada vez que se encuentre una letra "o".

```python
frase = input("introduce una frase: ")

for o in frase.lower():
    if "o" in o:
        print("'o' encontrada")
```

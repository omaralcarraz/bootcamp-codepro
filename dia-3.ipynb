Challenge 2 🤺 - 20 min
Crear una función que reciba dos letras (parámetros) y que verifique si son iguales:

Si son iguales, imprimir "Las letras son iguales".

De lo contrario, imprimir "Las letras son diferentes".

```python
letra_1=input("introduce la primer letra: ")
letra_2=input("introduce la segunda letra: ")

def compar_letras(letra1,letra2):

    if letra1.lower()==letra2.lower():
        print("las letras son iguales")
    else:
        print("las letras son diferentes")
compar_letras(letra_1, letra_2)
```

Pista: cantidad de letras de una palabra en python

```python
def comparacion_cant_letras(palabra,numero):
    if len(palabra)==numero:
        return True
    else:
        return False
    
resultado = comparacion_cant_letras("python", 5)
print(resultado)
```

lista = [['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']]

Crear una función que imprima una grilla (lista de listas), una línea por cada lista dentro de la lista principal.

Ejemplo del resultado esperado:

['h', 'o', 'l', 'i', 's']
['h', 'e', 'l', 'l', 'o']
['h', 'a', 'l', 'l', 'o']

```python
lista = [['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']]
def listas():
    for i in lista:
        print(i)
listas()
```

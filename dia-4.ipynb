Challenge 1 🤺
Escribir un programa que cree un diccionario vacío y lo vaya llenado con información sobre una persona (por ejemplo nombre, edad, sexo, teléfono, correo electrónico, etc.) que se le pida al usuario. 

Cada vez que se añada un nuevo dato debe imprimirse el contenido del diccionario.

```python
datos_cargados={}
print("A continuacion introduce los datos o 'salir' para terminar el programa")

while True:
    llave=input("Introduce el tipo de dato que vas a cargar")
    if llave.lower()== "salir":
        break
    valor=input(f"Introduce el valor de {llave}")
    datos_cargados[llave]=valor
    print(datos_cargados)
```

Challenge 2 🤺
Crear una tarjeta personal. El usuario (mediante input) puede elegir:

Crear datos nuevos
Modificar datos
Eliminar datos

```python
tarjeta_personal={
    "nombre":"omar",
    "telefono":"123456",
    "correo electronico":"sdsaf@fksanfa.com",
    "direccion":"mi casa"
    }

while True:
    print(tarjeta_personal)
    accion=input("Elige 'crear' 'reemplazar' o 'eliminar' datos, escribe 'salir' para terminar el programa")
    
    if accion.lower() == "crear":
        clave_utilizada=input(f"Elige el dato que quieres {accion}")
        nuevo_valor=input(f"introduce el valor de {clave_utilizada}")
        tarjeta_personal[clave_utilizada]=nuevo_valor
    elif accion.lower() == "reemplazar":
        clave_utilizada=input(f"Elige el dato que quieres {accion}")
        nuevo_valor=input(f"introduce el valor de {clave_utilizada}")
        tarjeta_personal[clave_utilizada]=nuevo_valor
    elif accion.lower() == "eliminar":
        clave_utilizada=input(f"Elige el dato que quieres {accion}")
        del tarjeta_personal[clave_utilizada]
    elif accion == "salir":
        break
    else:
        print("Opcion no valida")

```

Challenge 3 🤺
Crear una agenda de contactos. El usuario puede elegir:

Mostrar los contactos
Crear contacto
Modificar datos del contacto

```python
contactos={
    "omar":{"telefono":"1111", "mail":"omar@mail.com"}, 
    "juan":{"telefono":"222", "mail":"juan@mail.com"},
    "pedro":{"telefono":"333", "mail":"pedro@mail.com"},
}

while True:
    print(contactos)
    accion=input("Elige 'crear' o 'modificar' un contacto, escribe 'salir' para terminar el programa")
    if accion.lower()== "salir":
        break
    elif accion.lower() == "crear":
        contacto=input("introduce el nombre del contacto")
        telefono=input(f"introduce el telefono de {contacto}")
        mail=input(f"introduce el mail de {contacto}")
        contactos[contacto]={"telefono":telefono, "mail":mail}
    elif accion.lower() == "modificar":
        contacto=input("introduce el nombre del contacto que quieres modificar")
        telefono=input(f"introduce el nuevo telefono de {contacto}")
        mail=input(f"introduce el nuevo mail de {contacto}")
        contactos[contacto]={"telefono":telefono, "mail":mail}
    else:
        print("Opcion no valida")




```
